```python
import numpy as np

def random_V():
    """Gera uma amostra aleatoria para a variavel 
    V = min{k | X_1 + ... + X_k} >= 1, onde cada 
    X_i eh independente e possui uma distribuicao 
    uniforme entre 0 e 1."""

    k = 0
    Sk = 0
    while (Sk < 1):
        k += 1
        vk = np.random.rand()
        Sk += vk
    return k
```

```python
def estimateExpectedValue_V(n):
    """Estima o valor esperado da variavel 
    V = min{k | X_1 + ... + X_k} >= 1} a partir
    da media amostral de uma amostra de tamanho n,
    onde X_i eh independente e possui uma distribuicao
    uniforme entre 0 e 1."""

    Sk = 0
    for k in range(n):
        vk = random_V()
        Sk += vk
    Mn = Sk/n
    return Mn
```

```python
import seaborn as sns
sns.set_style("whitegrid")
from tqdm import tqdm
from matplotlib import pyplot as plt

sampleSize = list(range(0,10000+1,50))
estimations = [estimateExpectedValue_V(n+1) for n in tqdm(sampleSize)]
print(f"Last estimation ({sampleSize[-1]} samples): {estimations[-1]}")
sns.lineplot(x = sampleSize, y = estimations)
plt.xlabel("Sample size")
plt.ylabel("Estimation of E[V]")
```
