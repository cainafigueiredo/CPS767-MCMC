```python
# Quest√£o 6.1

import numpy as np
import seaborn as sns

def random_exponential(lmbda):
  u = np.random.rand()
  x = -np.log(1-u)/lmbda
  return x
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_style('whitegrid')

def exponentialPDF(x, lmbda):
    return lmbda * np.exp(-lmbda * x)

def plotFunction(func, a, b, n):
    x = np.array(range(n+1))/n * (b-a) + a
    y = np.array([func(k) for k in x])
    sns.lineplot(x=x, y=y, color = 'red')

lmbda = 2
plotFunction(lambda x: exponentialPDF(x, lmbda), 0, 6, 100)
n = 5000
sample = np.array([random_exponential(lmbda) for i in range(n)])
sns.histplot(x = sample, stat = 'density')
plt.legend(["PDF of X", "Density of sample"])
plt.xlabel('x')
plt.ylabel('f_X(x)')
```
