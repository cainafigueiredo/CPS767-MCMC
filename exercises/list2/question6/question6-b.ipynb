```python
# Questão 6.2

import numpy as np
import seaborn as sns

def random_pareto(alpha, x0):
  u = np.random.rand()
  power = np.log(x0) - np.log(1-u)/alpha
  x = np.exp(power)
  return x
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from tqdm import tqdm

sns.set_style('whitegrid')

def paretoPDF(x, alpha, x_0):
    return alpha*x_0**alpha/x**(alpha+1)

def plotFunction(func, a, b, n):
    x = np.array(range(n+1))/n * (b-a) + a
    y = np.array([func(k) for k in x])
    sns.lineplot(x=x, y=y, color = 'red')

alpha = 8
x0 = 2
plotFunction(lambda x: paretoPDF(x, alpha, x0), x0-0.0002, 10, 100)
n = 5000
sample = np.array([random_pareto(alpha, x0) for i in tqdm(range(n))])
sns.histplot(x = sample, stat = 'density')
plt.legend(["PDF of X", "Density of sample"])
plt.xlabel('x')
plt.ylabel('f_X(x)')
```
